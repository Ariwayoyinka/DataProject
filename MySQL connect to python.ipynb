```python
import mysql.connector
```

```python
mydb=mysql.connector.connect(host="localhost",user="root",password="Lonimi2022@",database="Marketdata")
```

```python
mycursor=mydb.cursor() #stores the cursor of the current database,a cursor allows you to iterate a ser of rows returned by a query
```

```python
mycursor.execute("select Volume, AdjClose from Googledata")
result = mycursor.fetchall
```

```python
Volume = []
AdjClose = []

for i in mycursor:
    Volume.append(i[0])
    AdjClose.append(i[1])
print("Volume of trade = ", Volume)
print("Adjusted price = ", AdjClose)
```

```python
# Visualizing Data using Matplotlib
plt.plot(AdjClose, Volume)
plt.xlabel("Adjusted closing price")
plt.ylabel("Voulume of trade")
plt.title("Impact of trade volume on Adjusted closing price")
plt.show()
```
